#Day-5 (Trigonometry with Coding)

```python
# degree radians conversion
import math

def deg_to_rad(deg):
  return deg * (math.pi / 180)

def rad_to_deg(rad):
    return rad * (180 / math.pi)

print(f"Radian conversion= {deg_to_rad(180)}")
print(f"Degree conversion= {rad_to_deg(math.pi/2)}")
```

```python
angles = [0, 30, 45, 60, 90]

for deg in angles:
    rad = deg_to_rad(deg)
    print(deg, "deg:",
          "sin =", math.sin(rad),
          "cos =", math.cos(rad),
          "tan =", math.tan(rad))
```

```python
# Unit Circle Geometry

def unit_circle(theta_deg):
    rad = deg_to_rad(theta_deg)
    return (math.cos(rad), math.sin(rad))

print(unit_circle(0))
print(unit_circle(90))
print(unit_circle(45))
```

```python
# sin²θ + cos²θ = 1

def verify_identity(theta_deg):
    rad = deg_to_rad(theta_deg)
    return math.sin(rad)**2 + math.cos(rad)**2

print(verify_identity(45))
```

```python
# tanθ = sinθ / cosθ

def tan_from_sin_cos(theta_deg):
    rad = deg_to_rad(theta_deg)
    return math.sin(rad) / math.cos(rad)

print(tan_from_sin_cos(45))
```

```python
# Inverse Trigonometry

print(math.degrees(math.asin(0.5)))
print(math.degrees(math.acos(0)))
print(math.degrees(math.atan(1)))
```

```python
# Hypotenuse

def hyp(a, b):
    return (a*a + b*b)**0.5

print(hyp(3, 4))
```

```python
def angle_from_sides(opp, adj):
    return math.degrees(math.atan(opp/adj))

print(angle_from_sides(3, 3))
```

```python
# Sinusoidal Functions
# PE(pos)=sin(pos),cos(pos)

def sinusoidal_encoding(x):
    return math.sin(x), math.cos(x)

print(sinusoidal_encoding(1))
print(sinusoidal_encoding(10))
```

```python
# Rotation of a point
# Rotating a point by θ
# x′=xcosθ−ysinθ     # y′=xsinθ+ycosθ

def rotate_point(x, y, theta_deg):
    rad = deg_to_rad(theta_deg)
    x_new = x*math.cos(rad) - y*math.sin(rad)
    y_new = x*math.sin(rad) + y*math.cos(rad)
    return (x_new, y_new)

print(rotate_point(1, 0, 90))
```
