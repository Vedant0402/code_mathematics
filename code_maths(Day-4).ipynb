Day-4 (Coordinate Geometry)

```python
# Define a point
P = (3, 4)
Q = (7, 1)

print("Point P:", P)
print("Point Q:", Q)
```

```python
# Distance between 2 points (x,y)
def distance(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return ((x2 - x1)**2 + (y2 - y1)**2)**0.5

P = (3, 4)
Q = (7, 1)
print("Distance between P and Q:", distance(P, Q))
```

```python
# Midpoint
def midpoint(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return ((x1 + x2) / 2, (y1 + y2) / 2)

print("Midpoint:", midpoint(P, Q))
```

```python
# Slope of a line
def slope(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    if x2 == x1:
        return "Undefined (vertical line)"
    return (y2 - y1) / (x2 - x1)

print("Slope:", slope(P, Q))
```

```python
# Equation of a line (y = mx + c)
def line_equation(p1, p2):
    m = slope(p1, p2)
    if m == "Undefined (vertical line)":
        return "x = " + str(p1[0])
    x1, y1 = p1
    c = y1 - m * x1
    return f"y = {m}x + {c}"

print("Line Equation:", line_equation(P, Q))
```

```python
# Angel between two points
import math

def angle_between_lines(m1, m2):
    tan_theta = abs((m1 - m2) / (1 + m1 * m2))
    theta = math.degrees(math.atan(tan_theta))
    return theta

print("Angle between lines with slopes 2 and -1:", angle_between_lines(2, -1))
```

```python
# Vector (2D)
def vector(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return (x2 - x1, y2 - y1)

print("Vector PQ:", vector(P, Q))
```

```python
# Vector Magnitude
def magnitude(v):
    x, y = v
    return (x**2 + y**2)**0.5

V = vector(P, Q)
print("Magnitude of PQ:", magnitude(V))
```

```python
# Dot Product
def dot_product(v1, v2):
    return v1[0]*v2[0] + v1[1]*v2[1]

A = (1, 2)
B = (3, 4)
print("Dot Product:", dot_product(A, B))
```

```python
# Angel between two vectors
def angle_between_vectors(v1, v2):
    dot = dot_product(v1, v2)
    mag1 = magnitude(v1)
    mag2 = magnitude(v2)
    cos_theta = dot / (mag1 * mag2)
    theta = math.degrees(math.acos(cos_theta))
    return theta

print("Angle between vectors A and B:", angle_between_vectors(A, B))
```
